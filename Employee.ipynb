```
emp = spark.read.format("csv").option("header", "true").option("inferSchema","true").load("dbfs:/FileStore/shared_uploads/akshaysoni460@gmail.com/employee_data.csv")
emp1=emp.drop("_c0")
emp1.show(10)
```

```
from pyspark.sql import Window
from pyspark.sql.functions import rank
emp2=emp1.withColumn("rank",rank().over(Window.partitionBy("country").orderBy("salary")))
print("Printing the dataframe df2")
emp2.show()
```

```
from pyspark.sql.functions import avg,count,sum,min,max
emp1.groupBy("country").agg(avg("salary"),sum("salary"),min("salary"),max("salary")).show()
```

```
emp1.select("name","age","salary","country").where("salary > 2000 and salary < 2500").groupBy("country").min("salary").alias("125").show()
```

```
from pyspark.sql.functions import desc,dense_rank
emp1.withColumn("rank",dense_rank().over(Window.partitionBy().orderBy(desc("salary")))).show()
```

```
from pyspark.sql.functions import rank,row_number
emp1.withColumn("rank_dense",row_number().over(Window.partitionBy().orderBy(desc("salary")))).show()
```
